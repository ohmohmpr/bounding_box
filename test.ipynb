```python
# import iou.py in the same directory

import numpy as np
from iou import *
import matplotlib.pyplot as plt
```

```python
# Create two line segments (2D)
line1 = np.array([[0, 0], [1, 1]])
line2 = np.array([[0, 1], [1, 0]])

# Plot the two lines
plt.plot(line1[:, 0], line1[:, 1], label='line1')
print("X: ",line1[:, 0])
print("Y: ",line1[:, 1])
plt.plot(line2[:, 0], line2[:, 1], label='line2')
```

```python
# Plot line1, line2, and their intersection
intersection = find_intersection(line1, line2)
if intersection:
    plt.plot(intersection[0], intersection[1], 'ro', label='intersection')
    plt.plot(line1[:, 0], line1[:, 1], label='line1')
    plt.plot(line2[:, 0], line2[:, 1], label='line2')
    plt.legend()
    plt.show()
```

```python
# Rotate the line1 by 90 degrees and translate it by (1, 1)
line1_transformed = transform_line_segment(line1, (1, 1), np.pi / 2)
# Plot the two lines
plt.plot(line1[:, 0], line1[:, 1], label='line1')
plt.plot(line1_transformed[:, 0], line1_transformed[:, 1], label='line1_transformed')
plt.plot(line2[:, 0], line2[:, 1], label='line2')
# plot with constant factor on the x-axis and y-axis
plt.axis('equal')
plt.title('Rotated and translated line segment by 90 degrees and (1, 1)')
plt.legend()
```
